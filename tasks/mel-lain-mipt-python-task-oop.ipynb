__Задание 2 (1 балл):__ Описать класс комплексных чисел, поддерживающий представление в полярном виде. Произвести перегрузку магических методов для основных арифметических операций (+, -, \*, /). Допускается использование модуля `math`.

```python
import math

class ComplexNumber:
    def __init__(self,real,imaginary):
        self.__real = real
        self.__imaginary = imaginary    
    def __repr__(self):
        return f'<{self.__real} + {self.__imaginary} * i>'
    def __neg__(self):
        return ComplexNumber(-self.__real,-self.__imaginary)
    def __module__(self):
        return self.__real ** 2 + self.__imaginary ** 2
    def __add__(self,other):
        return ComplexNumber(self.__real + other.__real, self.__imaginary + other.__imaginary)
    def __sub__(self,other):
        return self + - other
    def __mul__(self,other):
        real = self.__real * other.__real - self.__imaginary * other.__imaginary
        imaginary = self.__real * other.__imaginary + self.__imaginary * other.__real
        return ComplexNumber(real, imaginary)
    def __truediv__(self,other):
        denominator = other.__module__()
        real = (self.__real * other.__real + self.__imaginary * other.__imaginary) / denominator
        imaginary = (self.__imaginary * other.__real - self.__real * other.__imaginary) / denominator
        return ComplexNumber(real, imaginary)
    def polar(self):
        return (math.hypot(self.__real,self.__imaginary), math.atan2(self.__imaginary, self.__real))
    
nums = [ComplexNumber(5.0,3.0), ComplexNumber(-5.0,-3.0)]

print(nums[0])
print(nums[1])

sum_nums = nums[0] + nums[1]
sub_nums = nums[0] - nums[1]
mul_nums = nums[0] * nums[1]
div_nums = nums[0] / nums[1]

print(f'{nums[0]} + {nums[1]} = {sum_nums}')
print(f'{nums[0]} - {nums[1]} = {sub_nums}')
print(f'{nums[0]} * {nums[1]} = {mul_nums}')
print(f'{nums[0]} / {nums[1]} = {div_nums}')


polar = nums[0].polar()
print(polar)
```

```python
class Vector:
    def __init__(self, vector_values_list):
        if(type(vector_values_list) is not list):
            raise ValueError(f'Type of {vector_values_list} is not list')
        self.__values = vector_values_list
        self.__dimension = len(vector_values_list)
    def __getitem__(self,key):
        return self.__values[key]
    def __repr__(self):
        return f'{self.__values}'
    def module(self):
        return sum([self[i] ** 2 for i in range(0,self.__dimension)]) ** .5
    def __add__(self,other):
        if(self.__dimension != other.__dimension):
            raise ValueError(f'The dimension of the first vector({self.__dimension}) are not equal the dimension of the second vector({other.__dimension})')
        return [self[i] + other[i] for i in range(0,self.__dimension)]
    def __sub__(self,other):
        if(self.__dimension != other.__dimension):
            raise ValueError(f'The dimension of the first vector({self.__dimension}) are not equal the dimension of the second vector({other.__dimension})')
        return [self[i] - other[i] for i in range(0,self.__dimension)]
    def __mul__(self,other):
        if(self.__dimension != other.__dimension):
            raise ValueError(f'The dimension of the first vector({self.__dimension}) are not equal the dimension of the second vector({other.__dimension})')
        return sum([self[i] * other[i] for i in range(0,self.__dimension)])
    def cos(self, other):
        if(self.__dimension != other.__dimension):
            raise ValueError(f'The dimension of the first vector({self.__dimension}) are not equal the dimension of the second vector({other.__dimension})')
        return (self * other) / (self.module() * other.module())
    
first_vector = Vector([11,12,13])
second_vector = Vector([6,7,8])

print(first_vector)
print(second_vector)

sum_vectors = first_vector + first_vector
sub_vectors = first_vector - first_vector
scalar_mult_vectors = first_vector * second_vector
cos_between_vectors = first_vector.cos(second_vector)
print(sub_vectors)
print(f'{first_vector.module()}')
print(scalar_mult_vectors)
print(cos_between_vectors)
```

При этом приватные атрибуты-поля (начинающиеся в объявлении с "\_\_") должны иметь в декорируемом классе корректное имя.

В случае конфликтов имён импортируемый атрибут должен получить имя с суффиксом "\_new".

```python
def copy_class_attrs(decorator_class):
    def __decorate(cls):
        for attr_name in decorator_class.__dict__:
            if(not attr_name.startswith('__') and not attr_name.endswith('__')):
                if (not callable(getattr(decorator_class, attr_name)) and attr_name.startswith('__')):
                    attr_name.replace(f'_{decorator_class.name}__','__')
                    setattr(cls,attr_name,getattr(decorator_class,attr_name))
                if(hasattr(cls, attr_name)):
                    new_attr_name = f'_new{attr_name}'
                    setattr(cls,new_attr_name,getattr(decorator_class,attr_name))
                else:
                    setattr(cls,attr_name,getattr(decorator_class,attr_name))
        return cls
    return __decorate

class Example:
    a = 'str'
    __d = 10
    def __init__(self):
        self.__a = 0
        self.__str ='hello world'
    def __p(self):
        pass
    def func(self):
        print('hello world')
    def sum(self):
        print('sum')
       
@copy_class_attrs(Example)
class Example2:
    def __init__(self):
        pass
    def func():
        pass
    
example = Example2()    
for attr in Example2.__dict__:
    print(attr)
```
