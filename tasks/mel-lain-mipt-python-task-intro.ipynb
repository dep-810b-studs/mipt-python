__Задание 4 (0.5 балла):__ Определить, является ли введённая строка палиндромом (то есть одинаково читается с обеих сторон). Предложите как минимум три различных решения.

````python
import re

def task_04_func(s):
    if(type(s) == str):
        s = re.sub('[^A-Za-zА-Яа-я0-9]+', '', s).lower()
        middle = int(len(s) / 2)
        return s[0:middle] == s[::-1][0:middle]
    else:
        raise TypeError('Invalid type argument, must be string')

def task_04_func_02(s):
    if(type(s) == str):
        s = re.sub('[^A-Za-zА-Яа-я0-9]+', '', s).lower()
        if(len(s)==1 or len(s)==0):
            return True
        else:
            if(s[0]==s[-1]):
                return task_04_func_02(s[1:-1])
            else:
                return False
    else:
        raise TypeError('Invalid type argument, must be string')
        
def task_04_func_03(s):
    if(type(s) == str):
        s = re.sub('[^A-Za-zА-Яа-я0-9]+', '', s).lower()
        middle = int(len(s) / 2) 
        for i in range(0, middle):
            if(s[i]!=s[-i-1]):
                return False
        return True
    else:
        raise TypeError('Invalid type argument, must be string')        

print(task_04_func('ll'))
print(task_04_func('lol'))
print(task_04_func('kek'))
print(task_04_func('Аргентина манит негра'))
print(task_04_func('А роза упала на лапу Азора'))

print(task_04_func_03('ll'))
print(task_04_func_03('lol'))
print(task_04_func_03('kek'))
print(task_04_func_03('Аргентина манит негра'))
print(task_04_func_03('А роза упала на лапу Азора'))

print(task_04_func_02('ll'))
print(task_04_func_02('lol'))
print(task_04_func_02('kek'))
print(task_04_func_02('Аргентина манит негра'))
print(task_04_func_02('А роза упала на лапу Азора'))

print(task_04_func_03('ll'))
print(task_04_func_03('lol'))
print(task_04_func_03('kek'))
print(task_04_func_03('Аргентина манит негра'))
print(task_04_func_03('А роза упала на лапу Азора'))        
````
